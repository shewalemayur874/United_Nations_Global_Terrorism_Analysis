## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#1 copy dataset
#(Creating a Duplicate dataset here)
df_clean = df.copy()


#2 Handle missing values
#removing columns with more than 50% missing values
missing_percent = (df_clean.isnull().sum() / len(df_clean)) * 100
cols_to_drop = missing_percent[missing_percent > 50].index
df_clean.drop(columns=cols_to_drop, inplace=True)


#3 filling numerical columns with median (median used, not affected by extreme values better then mean)
num_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
df_clean[num_cols] = df_clean[num_cols].fillna(df_clean[num_cols].median())


#4 filling categorical columns (text columns) with mode
cat_cols = df_clean.select_dtypes(include=['object']).columns
for col in cat_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])


#5 Remove Duplicates (Deletes repeated records)
df_clean.drop_duplicates(inplace=True)

#6 Feature Selection (Keeping only important columns to reduce complexity)
important_cols = ['iyear', 'imonth', 'iday',
                  'country_txt', 'region_txt', 'city',
                  'attacktype1_txt', 'targtype1_txt', 'weaptype1_txt',
                  'nkill', 'nwound', 'gname', 'success']

df_clean = df_clean[important_cols]

#7 Handle Outlier with (IQR Method)
#find middle values
Q1 = df_clean['nkill'].quantile(0.25)
Q3 = df_clean['nkill'].quantile(0.75)
#Main Data Spread
IQR = Q3 - Q1

#Normal Range (to Create safe Zone)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Fix Outlier (if the value is too high will reduce it and if it is too low it will increase it)
df_clean['nkill'] = np.where(df_clean['nkill'] > upper_bound, upper_bound, df_clean['nkill'])
df_clean['nkill'] = np.where(df_clean['nkill'] < lower_bound, lower_bound, df_clean['nkill'])
#identified extreme values using the IQR method and adjusted them to a normal range to ensure accurate analysis.


````

````python
# Convert to integer (casualties should be whole numbers)
df_clean['nkill'] = df_clean['nkill'].round().astype(int)
df_clean['nwound'] = df_clean['nwound'].round().astype(int)


#8 Fetaure Engineering
#Create new column (total casualties)
df_clean['total_casualties'] = df_clean['nkill'] + df_clean['nwound']

#9 Data type correction (Convert year into int format)
df_clean['iyear'] = df_clean['iyear'].astype(int)

#10 Final Execution
print("Cleaned Dataset Shape:", df_clean.shape)
df_clean.info()
df_clean.head(20)
````
